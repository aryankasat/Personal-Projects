DATA TRANSFORMATIONS - transforming and normalizing the data for accurate comparisons, removing of outliers 

```python
#normalizing the data as different teams have different Played Tournaments
data['PlayedGamesperTournament'] = data['PlayedGames']/ data['PlayedTournaments']
data['WonGamesperTournament'] = data['WonGames']/data['PlayedTournaments']
data['DrawnGamesperTournament'] = data['DrawnGames'] / data['PlayedTournaments']
data['LostGamesperTournament'] = data['LostGames']/data['PlayedTournaments']


new_data = data.drop(columns=['PlayedTournaments','WonGames','DrawnGames','LostGames','PlayedGames'],axis=1)
```

```python
#calculating the number of years the team has played tournaments and sorting based on new teams
new_data['TeamLaunch'] = new_data['TeamLaunch'].apply(pd.to_numeric,errors = "coerce")
new_data['NumberofYears'] = 2024-new_data['TeamLaunch']
drop_data = new_data.drop(columns=['TeamLaunch'],axis=1)
sorted_data = drop_data.sort_values(by = 'NumberofYears',ascending=True)
```

```python
#transforming the data to percentages
sorted_data['WonPercent'] = (sorted_data['WonGamesperTournament']/sorted_data['PlayedGamesperTournament']) * 100
sorted_data['DrawnPercent'] = (sorted_data['DrawnGamesperTournament']/sorted_data['PlayedGamesperTournament']) * 100
sorted_data['LostPercent'] = (sorted_data['LostGamesperTournament']/sorted_data['PlayedGamesperTournament']) * 100

percent_data = sorted_data.drop (columns=['WonGamesperTournament','DrawnGamesperTournament','LostGamesperTournament','PlayedGamesperTournament'])
```

```python
#removing of outliers based on the score, basket scored, basket given individually as these would be the teams which would be rather old teams which are mostly bought by other 
# companies
mean_score = percent_data['Score'].mean()
std_score = percent_data['Score'].std()

mean_Basketscored = percent_data['BasketScored'].mean()
std_Basketscored = percent_data['BasketScored'].std()

mean_Basketgiven = percent_data['BasketGiven'].mean()
std_Basketgiven = percent_data['BasketGiven'].std()


percent_data['Z_Score'] = (percent_data['Score']-mean_score)/std_score
percent_data['Z_BasketScored'] = (percent_data['BasketScored'] - mean_Basketscored)/ std_Basketscored
percent_data['Z_Basketgiven'] = (percent_data['BasketGiven'] - mean_Basketgiven) / std_Basketgiven

cleaned_data = percent_data[(np.abs(percent_data['Z_Score'])<=3) & (np.abs(percent_data['Z_BasketScored'])<=3) & (np.abs(percent_data['Z_Basketgiven'])<=3)]

cleaned_data = cleaned_data.drop(columns=['Z_Score','Z_BasketScored','Z_Basketgiven'],axis=1)
```

```python
#finding whether there are any outliers which might cause disturbance in analysis
cols = ['Score', 'BasketScored', 'BasketGiven',
       'WonPercent', 'DrawnPercent', 'LostPercent']
cleaned_data[cols].skew()
```

FEATURE ENGINEERING - selection of most relevant features for performing the analysis

```python
#selecting the most relevant features and choosing the optimal number of components from the entire data
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler

cols = ['Score', 'BasketScored', 'BasketGiven',
       'TournamentChampion', 'Runnerup', 'HighestPositionHeld',
       'NumberofYears', 'WonPercent', 'DrawnPercent', 'LostPercent']

scaler = StandardScaler()
standardised_data = scaler.fit_transform(cleaned_data[cols])

pca = PCA()
principal_data = pca.fit_transform (standardised_data)

explain_variance = pca.explained_variance_ratio_.cumsum()

n_components = np.argmax (explain_variance>=0.95)+1

pca = PCA (n_components= n_components)
reduced_data = pca.fit_transform(standardised_data)
loadings = pca.components_

loadings_df = pd.DataFrame (loadings,columns = cols,index = ['PCA1','PCA2','PCA3','PCA4'])

loadings_df
```

```python
#finding the sum of each column for the loadings across each principle component
loadingssum = loadings_df.sum()
print (loadingssum)
```
